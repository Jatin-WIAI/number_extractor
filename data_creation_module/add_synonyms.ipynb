```python
import pandas as pd
import json
```

```python
files = ["/home/jatin/number_extractor/util_data/hi_frac_dict.json"]
```

```python
words_to_add = {
    "उनतालीस":"उनचालीस",
    "तैंतालीस":"तरालीस",
    "चौंतालीस":"चौआलिस",
    "उनचास":"उनपचास",
    "इक्यावन":"इक्कावन",
    "छः":"छे",
    "सौ": "सो"
}
```

```python
words_to_find =list(words_to_add.keys())
for file in files:
    with open(file,"r") as f:
        data = json.load(f)
    new_data = {}
    count=0
    for k,v in data.items():
        new_v = []
        for item in v:
            new_v.append(item.lstrip().rstrip())
            # if count<50:
            #     print(item)
            for word in words_to_find:
                word = " "+word.lstrip().rstrip()+" "
                item = " "+item+" "
                print(k,word,item)
                if item.find(word)!=-1:
                    new_v.append(item.replace(word," "+words_to_add[word.lstrip().rstrip()]+" ").lstrip().rstrip())
        count+=1
        new_data[k]=list(set(new_v))
    with open(file,"w") as f:
        json.dump(new_data,f,ensure_ascii=False)
    # print(new_data)
    # break
        
```
