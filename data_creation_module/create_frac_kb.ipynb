```python
import json
from num_to_words import num_to_word
```

```python
lang = "mr"
```

```python
decimals = range(0,10)
numbers = range(0,100)

frac_dict = {}
for number in numbers:
    for decimal in decimals:
        syn=[]
        if decimal ==5:
            if number not in [0,1,2]:
                word = num_to_word(number,lang).replace(",","")
                syn+=["साड़े "+word, "साढ़े "+word, "साढे "+word, "साडे "+word]
            if number == 0:
                syn+=["अर्धा","अर्ध"]
            if number == 1:
                syn+=["देढ", "डेढ", "दिढ", "दिड"]
            if number == 2:
                syn+=["अडीच"]
        else:
            num_word = num_to_word(number,lang).replace(",","")
            deci_word = num_to_word(decimal,lang).replace(",","")
            syn+=[num_word+" "+"पूर्णांक "+deci_word ]
        frac_dict[str(number)+"."+str(decimal)] = syn
```
