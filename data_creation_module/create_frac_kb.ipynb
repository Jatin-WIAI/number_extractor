```python
decimals = range(0,10)
numbers = range(0,100)

frac_dict = {}
for number in numbers:
    for decimal in decimals:
        syn=[]
        if decimal ==5:
            if number not in [0,1,2]:
                word = num_to_word(number,lang).replace(",","")
                syn_list = lan_kb_dict[lang]["half_with_more_than_2"]
                syn+=[x+" "+word for x in syn_list]
            if number == 0:
                syn+=lan_kb_dict[lang]["half"]
            if number == 1:
                syn+=lan_kb_dict[lang]["one_and_half"]
            if number == 2:
                syn+=lan_kb_dict[lang]["two_and_half"]
        else:
            num_word = num_to_word(number,lang).replace(",","")
            deci_word = num_to_word(decimal,lang).replace(",","")
            syn+=[num_word+" "+lan_kb_dict[lang]["point"] +" "+deci_word ]
        frac_dict[str(number)+"."+str(decimal)] = syn

numbers = range(150,10000,100)
for number in numbers:
    syn=[]
    if number == 150:
        syn+=lan_kb_dict[lang]["one_fifty"]
    elif number == 250:
        syn+=lan_kb_dict[lang]["two_fifty"]
    else:
        syn_list = lan_kb_dict[lang]["more_than_two_fifty"]
        # print((number-50)/100)
        syn+=[x.replace("NUMBER",num_to_word(int((number-50)/100),lang).replace(",","")) for x in syn_list]
    frac_dict[str(number)] = syn

numbers = range(1500,100000,1000)
for number in numbers:
    syn=[]
    if number == 1500:
        syn+=lan_kb_dict[lang]["one_thousand_five_hundred"]
    elif number == 2500:
        syn+=lan_kb_dict[lang]["two_thousand_five_hundred"]
    else:
        syn_list = lan_kb_dict[lang]["more_than_two_thousand_five_hundred"]
        syn+=[x.replace("NUMBER",num_to_word(int((number-500)/1000),lang).replace(",","")) for x in syn_list]
    frac_dict[str(number)] = syn
```
