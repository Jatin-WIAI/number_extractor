```python
file_path = "/home/jatin/number_extractor/util_data/hi_num_word_dict.json"
import json
# Read the json file
with open(file_path,"r") as f:
    data = json.load(f)
```

```python
from word2number import w2n
```

```python
new_data = {}
for k,v in data.items():
    try: 
        num = w2n.word_to_num(k)
        new_data[num] = v
    except:
        new_data[k] = v
```

```python
with open("/home/jatin/number_extractor/util_data/hi_num_word_dict_new.json","w") as f:
    json.dump(new_data,f, indent=4, ensure_ascii=False)
```
