## Remove accents


```python
import json
with open("/home/jatin/huggingface_demo/number_extractor/util_data/accents_list.json","r") as f:
    accent_dict = json.load(f)["hi"]
```

```python
files = ['/home/jatin/huggingface_demo/number_extractor/util_data/hi_deci_keywords.json',
 '/home/jatin/huggingface_demo/number_extractor/util_data/hi_frac_keywords.json',
 '/home/jatin/huggingface_demo/number_extractor/util_data/hi_wh_num_dict.json']
```

```python
for file in files:
    with open(file,"r") as f:
        data = json.load(f)
    new_data = {}
    for k,v in data.items():
        new_data[k]=v
        for x in v:
            for acc,rep in accent_dict.items():
                x = x.replace(acc,rep)
            new_data[k]+=x
```

```python
for file in files:
    with open(file,"r") as f:
        data = json.load(f)
```

```python
for file in files:
    with open(file,"r") as f:
        data = json.load(f)
        new_data = data
        for k,v in data.items():
            new_v = []
            for x in v:
                for acc,rep in accent_dict.items():
                    x = x.replace(acc,rep)
                new_v.append(x)
            new_data[k] = list(set(v+new_v))
    with open(file[:-5]+"_deacentifined.json","w") as f:
        json.dump(new_data,f,ensure_ascii=False)
```
